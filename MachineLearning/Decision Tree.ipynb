## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
df['Embarked']=df['Embarked'].fillna(df['Embarked']=='S')
df
#print(df.isnull().sum())
```

```python
#Impute the missing values in Fare column with median value
df['Fare']=df['Fare'].fillna(df['Fare'].median())
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
df['Age']=df['Age'].fillna(df['Age'].median())
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
df['FamilySize'] =df['Parch']+df['SibSp']+1 
df
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
import numpy as np
df['IsAlone'] = np.where(df['FamilySize'] == 1, True, False)
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
s=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
s
```

## Model Building

```python
#Prepare X and y
x=df.values[:, 1:10]
y=df.values[:, 0]
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
s['Sex']=label.fit_transform(s['Sex'])
s['IsAlone']=label.fit_transform(s['IsAlone'])
#s['Embarked_Q']=label.fit_transform(s['Embarked_Q'])
#s['Embarked_S']=label.fit_transform(s['Embarked_S'])
#s=s.drop(columns=['Embarked_C'])
t=pd.get_dummies(s,columns=['Embarked'],drop_first=True).replace({True:1,False:0})
t
```

```python
# concat df_categorical with original df
#t.columns
t
```

```python
#Do a pairplot of all features
sns.pairplot(t)
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
x=t[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','FamilySize', 'IsAlone','Embarked_Q', 'Embarked_S']]
y=t['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier,plot_tree
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
#Print the tree

# Importing required packages for visualization
import seaborn as sns
plot_tree(model, feature_names=x.columns, class_names=["0", "1"], filled=True)
plt.show()
```

```python
# plotting tree with max_depth=3
plot_tree(model, feature_names=x.columns, class_names=["0", "1"], filled=True,max_depth=3)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
y_pred = model.predict(x_test)  # Predictions on validation set
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Precision
precision = precision_score(y_test, y_pred)
print(f"Precision: {precision:.4f}")

# Recall
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall:.4f}")

# F1 Score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")
```
